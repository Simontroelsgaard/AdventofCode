```python
import numpy as np
f = open("C:/Users/stroelsg/OneDrive - Capgemini/Desktop/adventofcode/input8.txt", "r")
L = []
for i in f:
    L.append(i.split()[0])
n = len(L[0])
m = len(L)
Trees = np.zeros([n,m])
Vis = np.zeros((n,m))
for i in range(n):
    for j in range(m):
        Trees[i,j] = L[i][j]
h = -np.ones(4)
for i in range(n):
    h = -np.ones(4)
    for j in range(m):
        if  Trees[i,j] > h[0]:
            Vis[i,j] = 1
            h[0] = Trees[i,j]
        if Trees[i,n-1-j] > h[1]:
            Vis[i,n-1-j] = 1
            h[1] = Trees[i,n-1-j]
        if Trees[j,i] > h[2]:
            Vis[j,i] = 1
            h[2] = Trees[j,i]
        if Trees[n-1-j,i] > h[3]:
            Vis[n-1-j,i] = 1
            h[3] = Trees[n-1-j,i]     

print((np.sum(Vis)))
```

```python
# Find score of trees
def calscore(Vis,Trees,i,j,n):
    h = Trees[i,j]
    m = np.zeros(4)
    for k in range(1,n):
        if i+k == n:
            m[0] = k-1
            break
        if Trees[i+k,j] >= h:
            m[0] = k
            break
    for k in range(1,n):
        if j+k == n:
            m[1] = k-1
            break
        if Trees[i,j+k] >= h:
            m[1] = k
            break
    for k in range(1,n):
        if i-k == -1:
            m[2] = k-1
            break
        if Trees[i-k,j] >= h or i-k == 0:
            m[2] = k
            break
    for k in range(1,n):
        if j-k == -1:
            m[3] = k-1
            break
        if Trees[i,j-k] >= h or j-k == 0:
            m[3] = k
            break
    score = m[0]*m[1]*m[2]*m[3]
    return score,m
maks = 0
for i in range(n):
    for j in range(m):
        s,mm = calscore(Vis,Trees,i,j,n)
        if s > maks:
            maks = s
print(maks)
```
