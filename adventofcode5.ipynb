```python
# Second part
f = open("C:/Users/stroelsg/OneDrive - Capgemini/Desktop/adventofcode/input5.txt", "r")
n = 9
Lstart = [[] for _ in range(n)]
count = 0

for i in f:
    if i in ['\n', '\r\n']:
        break
    c = 0
    spli = i.split(' ')
    for j in range(n-1):
        if spli[j] == '':
            del spli[j],spli[j],spli[j]
            
    Lstart[n-1-count].append(spli)
    
    print(i)
    count +=1

# Build lists with containers
L = [[] for _ in range(n)]
for i in range(0,n):
    for j in range(1,n):
        if Lstart[j][0][i] != '':
            L[i].append(Lstart[j][0][i])

print(L,'L')
# Move creates 
for i in f:
    s = i.split()
    numtomove = int(s[1])
    tos = int(s[5])-1
    froms = int(s[3])-1
    print(numtomove,tos,froms,len(L[froms]),L[froms],i,'\n','L',L)
    if numtomove > len(L[froms]):
        for j in range(len(L[froms])):
            L[tos].append(L[froms][j])
        for j in range(len(L[froms])):
            L[froms].pop()
    else:
        for j in range(numtomove):
            L[tos].append(L[froms][len(L[froms])-numtomove+j])
        for j in range(numtomove):
            L[froms].pop()
        
for i in range(n):
    print(L[i])
```
