```python
def Tailmove(K):
    for k in range(0,9):
        Hid = np.array([K[k,0],K[k,1]])
        Tid = np.array([K[k+1,0],K[k+1,1]])
        Copy = np.copy(Tid)
        I = Hid-Tid
        #print(I,'I')
        D = np.sum(I**2)
        #print(Tid,'Tid',I,D,'Hid',Hid)
        if I[0] > 0 and I[1] > 0 and D > 3:
            Tid += np.array([1,1])
        if I[0] < 0 and I[1] < 0 and D > 3:
            Tid += np.array([-1,-1])
        if I[0] < 0 and I[1] > 0 and D > 3:
            Tid += np.array([-1,1])
        if I[0] > 0 and I[1] < 0 and D > 3:
            Tid += np.array([1,-1])
        if I[0] == 0 and (I[1] > 1 or I[1] < -1):
            Tid[1] += np.sign(I[1])
        if I[1] == 0 and (I[0] > 1 or I[0] < -1):
            Tid[0] += np.sign(I[0])
        K[k+1,:] = Tid
        #print(Tid,Hid)
        if (Tid == Copy).all():
            print('break')
            break
    #print(Tid,'Tid')
    return K
```

```python
def moveknot(K,I,Ar):
    k = 0
    n = 9
    #print(K,'K')
    if I[0] == 'U':
        for j in range(0,int(I[1])):
            K[k] = K[k] + np.array([-1,0])
            
            K = Tailmove(K)
            Ar[int(K[n,0]),int(K[n,1])] = 1
    if I[0] == 'R':
        for j in range(0,int(I[1])):
            K[k] = K[k] + np.array([0,1])
            K = Tailmove(K)
            Ar[int(K[n,0]),int(K[n,1])] = 1
    if I[0] == 'D':
        for j in range(0,int(I[1])):
            K[k] = K[k] + np.array([1,0])
            K = Tailmove(K)
            Ar[int(K[n,0]),int(K[n,1])] = 1
    if I[0] == 'L':
        for j in range(0,int(I[1])):
            K[k] = K[k] + np.array([0,-1])
            K = Tailmove(K)
            Ar[int(K[n,0]),int(K[n,1])] = 1

    return K,Ar

```
