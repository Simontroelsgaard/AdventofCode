```python
#Saras solution
COIs = [20, 60, 100, 140, 180, 220]
SSs = []
c = 0
X = 1

visible = [[0] * 40]*6

def draw_sprite(X,c):
    row = c // 40
    rowpos = c-(row*40)
    if rowpos in [X-1,X,X+1]:
        print('#',end=' ')
    else:
        print('.',end=' ')

with open("C:/Users/stroelsg/OneDrive - Capgemini/Desktop/adventofcode/input10.txt") as f:
    for line in f.readlines():
        parsed = line.strip().split()
        if parsed[0] == 'addx':
            draw_sprite(X,c)
            c += 1
            # check if meets COIS
            if c in COIs:
                SSs.append(c*X)
            
            if c%40 == 0:
                print()
            
            draw_sprite(X,c)
            c += 1
            X += int(parsed[-1])
        else: #noop
            draw_sprite(X,c)
            c += 1
        if c in COIs:
            SSs.append(c*X) 
        if c%40 == 0:
            print()
        
            
print(sum(SSs))
```
